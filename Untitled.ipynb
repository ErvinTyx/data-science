```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a count plot for the 'Bankrupt?' column
plt.figure(figsize=(6, 4))
sns.countplot(x='Bankrupt?', data=data, hue='Bankrupt?',palette='viridis', legend=False)
plt.title('Frequency of Bankrupt')
plt.xlabel('Bankrupt?')
plt.ylabel('Count')
plt.show()
```

```python
data_amount_skewness = data.skew(numeric_only=True)
print(data_amount_skewness)
```

```python
# Create a figure with two subplots
plt.figure(figsize=(10, 4))

# Plot KDE plot (Kernel Density Estimate)
plt.subplot(1, 2, 1)
sns.kdeplot(x=' Operating Gross Margin', data=data, fill=True, color='blue')
plt.title('Kernel Density Estimate Plot of Operating Gross Margin')
plt.xlabel('Operating Gross Margin')
plt.ylabel('Density')

# Plot Count plot
plt.subplot(1, 2, 2)
sns.countplot(x=' Operating Gross Margin', data=data, hue=' Operating Gross Margin', palette='viridis', legend=False)
plt.title('Count Plot of Operating Gross Margin')
plt.xlabel('Operating Gross Margin')
plt.ylabel('Count')

# Display the skewness value
plt.figtext(0.7, 0.8, f"Skewness: {data_amount_skewness[' Operating Gross Margin']:.2f}",fontsize=10, ha='center')
plt.tight_layout()
plt.show()
```

```python
loan_binned1, bins1  = np.histogram(data[' Operating Gross Margin'],bins = 10)
print(loan_binned1)
print(bins1)
```

```python
loan_binned1,bins1 = np.histogram(data[' Operating Gross Margin'], bins=10)
print(loan_binned1)
print(bins1)
plt.figure(figsize=(12, 3))
sns.histplot(data[' Operating Gross Margin'], bins=bins1, edgecolor='black', alpha=0.7, kde=True)
plt.title('Histogram of LoanAmount with 10 Bins')
plt.xlabel(' Operating Gross Margin')
plt.ylabel('Frequency')
plt.grid(True)

mean_price = data[' Operating Gross Margin'].mean()
median_price = data[' Operating Gross Margin'].median()
mode_price = data[' Operating Gross Margin'].mode().squeeze() #convert it into a Series
std = data[' Operating Gross Margin'].std()
# Add vertical lines at the position of mean, median, mode
plt.axvline(mean_price, color='red', label="Mean")
plt.axvline(median_price, color="black", label="Median")
plt.axvline(mode_price, color="green", label="Mode")
plt.axvline(std, color="yellow", label="Std")

plt.legend();
plt.show()

print('mean:',mean_price, 'med:', median_price, 'mode:', mode_price, 'std:',std)
```
